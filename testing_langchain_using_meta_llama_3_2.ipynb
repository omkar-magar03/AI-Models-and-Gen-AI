```python
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline
from langchain.prompts import PromptTemplate
```

```python
# for not printing warning messages while showing the output
from transformers.utils.logging import set_verbosity_error
set_verbosity_error()
```

```python
# checking if GPU is available
import torch
print(torch.cuda.is_available())
```

```python
# downloading model
model = pipeline(
    task = "text-generation",
    model = "meta-llama/Llama-3.2-1B",
    max_length = 256,
    truncation = True
)
```

```python
# creating llm by wrapping model into HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=model)
```

```python
# prompt template
template = PromptTemplate.from_template(
    "Please explain {topic} topic to me. Use {sources} as reference. And do not copy-paste from the wikipedia."
)
```

```python
# taking input from user
topic = input("What topic would you like know: ")
sources = input("Sources: ")
```

```python
# creating a langchain chain
chain = template | llm      # takes templates -> fills it up with correct variables -> passes it LLM
```

```python
# executing the chain
response = chain.invoke({
    "topic" :  topic,
    "sources" : sources
})
```
